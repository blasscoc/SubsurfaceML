```python
from os import path 
from PIL import Image
from h5py import File as H5File
import numpy as np
import matplotlib.pylab as plt
```

```python
import matplotlib.colors as colors
facies_colors = ['#FFFFFF', '#FF0800', '#50C878','#FADA5E', '#0080FF', 
                 '#FC0FC0', '#F9A602', '#A0A0AA', '#000000']
facies_names= ['N/A', 'Basement', 'Growth1', 'Prograde1', 'Prograde2', 'cont1', 'mtc', 'cont2']
cmap_facies = colors.ListedColormap(facies_colors[0:len(facies_colors)], 'indexed')
```

```python
DATAPATH = path.expanduser('~/Documents/Hackathon/AgileSprint2018/PoseidonCore/PharosWhiteLight/')
```

```python
label_data = path.join(DATAPATH, 'PHAROS-1_WL_1m_Core1_4963_00m_to_4964_00m.h5')
core_image = path.join(DATAPATH, 'PHAROS-1_WL_1m_Core1_4963_00m_to_4964_00m.jpg')

with H5File(label_data) as fp:
    labels = fp['WL_1m_Core1_4963_00m_to_4964_00m'][:]
    
    # better as a numpy array
core_image = np.array(Image.open(core_image))
```

```python
plt.figure(figsize=(50,26))
plt.imshow(core_image)
plt.imshow(labels[:,:,0], alpha=0.5, cmap=cmap_facies, vmin=0, vmax=9)
plt.show()
```
